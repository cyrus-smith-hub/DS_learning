```python
#Задание 1+2

#Вам нужно помочь секретарю автоматизировать работу.
#Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

#не стал переделывать под процедурный стиль, чтобы не поломать. пока так.

#---data
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}
#---


def say_hi():
  print('=== Програма - секретарь ===')
  print('Введите команду из представленных ниже и нажмите Enter..')



def command_help():
  print()
  print('p - узнать владельца документа по его номеру')
  print('s - узнать по номеру документа полку, на которой он хранится')
  print('l - просмотр полной информации по всем документам')
  print('ads - добавить новую полку, просмотр всех полок')
  print('ds - удалить полку (только если она пустая), просмотр всех полок')
  print('ad - добавить новый документ в БД и на полку')
  print('d - удалить документ из БД')
  print('m - переместить документ на другую полку')
  print('h - показать набор команд')
  print('q - выход')
  print('doc - документация на функции')



def press_p():
  """
  Обработка ввода пользователем команды 'p'
  Показывает имя владельца по номеру документа
  """
  number = input('введите номер документа: ')
  answer = name_from_number(number)

  if answer:
    print(f">> данные владельца документа: {answer}")
  else:
    print('>> по данному номеру данных не найдено')



def name_from_number(number):
  """
  Поиск имени владельца документа по номеру документа
  Аргументы: номер документа, str
  Возвращает: данные по ключу 'name' из БД документов, str
  Если по номеру документа данных нет, возвращает False 
  """
  #print('name_from_number is working with number: ', number)

  answer = False #по-умолчанию возвращаем "не найдено"

  for el in documents:
    if str(number) == el['number']:
      answer = el['name']
  return answer



def press_s():
  """
  Обработка ввода пользователем команды 's'
  Аргументы: нет
  Возвращает: None
  Показывает полку, на которой хранится документ, по номеру документа
  """
  number = input('введите номер документа: ')
  answer = shelf_from_number(number)
  
  if answer:
    print(f'>> документ находится на полке {answer}')
  else:
    print('>> по данному номеру данных не найдено')



def shelf_from_number(number):
  """
  Поиск номера полки по номеру документа
  Аргументы: номер документа, str
  Возвращает: номер полки, str
  Если по номеру документа данных нет, возвращает False 
  """
   
  answer = False

  for el, item in directories.items():
    #print(el, item)
    if number in item:
      answer = el
    
  return answer



def press_l():
  """
  Обработка ввода пользователем команды 'l'
  Аргументы: нет
  Возвращает: None
  Показывает список всех документов с полками, на которых каждый из них хранится
  """
  print('Текущий список документов:')
  for el in documents:
    print(f">> № {el['number']}, тип: {el['type']}, владелец: {el['name']}, полка хранения: {shelf_from_number(el['number'])}")



def press_ads():
  """
  Обработка ввода пользователем команды 'ads'
  Аргументы: нет
  Возвращает: None
  Добавляет новую полку
  """
  shelf_number = input('введите номер новой полки: ')

  if shelf_number not in directories:
    directories[shelf_number] = []
    print('>> полка добавлена')
  else:
    print('>> полка с таким номером существует. ', end='')

  print('>> Всего полок: ', ', '.join(directories.keys()) + '.')
  


def press_ds():
  """
  Обработка ввода пользователем команды 'ds'
  Аргументы: нет
  Возвращает: None
  Удаляет указанную полку, только если она вообще существует и на ней ничего нет
  """
  shelf_number = input('введите номер полки для удаления: ')

  if shelf_number not in directories.keys():
    print('>> такой полки нет. ', end = '')

  elif shelf_number in directories.keys():

    if directories[shelf_number] != []:
      print('>> полка не пуста. ', end = '')
    else:
      directories.pop(shelf_number)
      print('>> полка удалена. ', end = '')

  print('Всего полок: ', ', '.join(directories.keys()) + '.')



def press_ad():

  """
  Обработка ввода пользователем команды 'ad'
  Аргументы: нет
  Возврат: None
  Обеспечивает добавление документа в БД и сразу на полку, если такая полка есть
  В конце раобты отображает список всех документов
  """
  stop_word = 'q'

  print(f'для отмены введите {stop_word} в любом поле')
  doc_number = input('введите номер документа:')
  if doc_number == stop_word: return

  doc_type = input('введите тип документа: ')
  if doc_type == stop_word: return

  name = input('введите владельца документа: ')
  if name == stop_word: return

  shelf_number = input('введите номер полки для документа: ')
  if shelf_number == stop_word: return
  
  if shelf_number not in directories.keys():
    print('>> такой полки нет, ввод придется повторить')
  
  else:
    documents.append({'type':doc_type, 'number':doc_number, 'name': name})
    directories[shelf_number].append(doc_number)
    print('>> документ добавлен')
  
  press_l()



def press_d():
  """
  Обработка ввода пользователем команды 'd'
  Аргументы: нет
  Возврат: None
  Обеспечивает удаление документа из БД и сразу с полки, если документ есть в БД
  В конце раобты отображает список всех документов
  """
  doc_number = input('введите номер документа:')
    
  for i in range(len(documents)):
    if documents[i]['number'] == doc_number:
      documents.pop(i) #удалили документ из списка документов
      directories[shelf_from_number(doc_number)].remove(doc_number) #удалили с полки

      print('>> документ удален')
      press_l()
      return
    else:
      doc_exist = False

  if doc_exist == False:
    print('>> документ не найден')
    press_l()



def press_m():
  """
  Обработка ввода пользователем команды 'm'
  Аргументы: нет
  Возврат: None
  Обеспечивает перемещение документа на другую полку
  В конце раобты отображает список всех документов  
  """
  stop_word = 'q'
  doc_exist = False
  
  print(f'для отмены введите {stop_word} в любом поле')

  doc_number = input('введите номер документа:')
  if doc_number == stop_word: return

  new_shelf = input('введите номер полку, на которую переместить документ: ')
  if doc_number == stop_word: return

  for i in range(len(documents)):
    if documents[i]['number'] == doc_number: #всегда проверяем наличие документа в БД документов для надежности
      doc_exist = True

  if doc_exist == True and new_shelf in directories: #если есть и документ и полка
    if shelf_from_number(doc_number) == new_shelf: #если он уже там, не нужно ничего делать
      print(f'>> документ уже на полке {new_shelf}, перемещение не требуется')
      press_l()
      return

    directories[shelf_from_number(doc_number)].remove(doc_number) #удалили со старой полки
    directories[new_shelf].append(doc_number) #добавили на новую полку
    print('>> документ перемещен')
    press_l()

  if doc_exist == False:
    print('>> документ не найден')
    press_l()

  if new_shelf not in directories.keys():
    print('>> такой полки не существует')
    press_l()


def press_doc():
  """
  Просмотр документации на все функции, использованные в работе
  """
  print('Документация на функции:', end =  '\n\n')
  print('press_p',press_p.__doc__)
  print('name_from_number',name_from_number.__doc__)
  print('press_s',press_s.__doc__)
  print('shelf_from_number', shelf_from_number.__doc__)
  print('press_l', press_l.__doc__)
  print('press_ads', press_ads.__doc__)
  print('press_ds', press_ds.__doc__)
  print('press ad', press_ad.__doc__)
  print('press d', press_d.__doc__)
  print('press_m', press_m.__doc__)
  print('press_doc', press_doc.__doc__)



def main():

  while True:
    print()
    user_command = input("введите команду (помощь - 'h'): ")

    if user_command == 'q':
      print('>> Выход')
      break

    elif user_command == 'h':
      command_help()

    elif user_command == 'p':
      press_p()

    elif user_command == 's':
      press_s()

    elif user_command == 'l':
      press_l()

    elif user_command == 'ads':
      press_ads()

    elif user_command == 'ds':
      press_ds()

    elif user_command == 'ad':
      press_ad()

    elif user_command == 'd':
      press_d()

    elif user_command == 'm':
      press_m()

    elif user_command == 'doc':
      press_doc()

    else:
      print('>> команда не найдена..')


#---body

say_hi()
command_help()
main()
```
