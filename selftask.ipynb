```python
#Генерация пароля
#Заданы длина, использование регистров, символов и чисел


class MyPass:
    """
    Генерация пароля с заданными параметрами (len, l, u, s, n)
    Для генерации пароля сначала генерируется шаблон вида (пример):
    
    lusnnln
    
    где:
    l - low_char
    u - upper_vhar
    s - simbol
    n - number
    
    Каждый элемент шаблона это обозначение вида требуемого символа
    Шаблон генерируется на основе случайной выборки
    Шаблон проверяется на наличие всех требуемых элементов
    Если шаблон не удовлетворяет требованиям - генерация повторяется
    Пароль генерируется заменой в шаблоне символов типа елементов
    На случайные элементы из соответствующих строк:
    
    aR=45t3
    """
    
    
    def __init__(self,length = 8, use_low = True, use_upper = True, use_simb = True, use_num = True):
        self.length = length
        self.use_low = use_low
        self.use_upper = use_upper
        self.use_simb = use_simb
        self.use_num = use_num
        
        self.pass_template = ''
        self.el_for_template = ''
        
        self.usage_char = 'qwertyuiopasdfghjklzxcvbnm'
        self.usage_num = '0123456789'
        self.usage_simb = '_!@#$%.:,^&*()=;+/'
        
        self.counter = 0 #количество итерация для генерации шаблона с требуемыми параметрами
        
        
    def pass_gen(self):
        """
        Генерация пароля по ранее сгенерированному шаблону
        На место каждого символа из шаблона вставляется случайный элемент соответствующего типа
        """
        my_pass = ''
        for el in self.pass_template_gen():
            if el == 'l': my_pass += self.get_rand_low()
            if el == 'u': my_pass += self.get_rand_upper()
            if el == 's': my_pass += self.get_rand_simb()
            if el == 'n': my_pass += self.get_rand_num()

        return my_pass
    
    
    def pass_template_gen(self):
        """
        Генерация шаблона пароля на основе ранее сгенерированного списка заданных элементов
        Шаблон проверяется на соответствие требованиям по составу
        В шаблоне есть хотя бы один элемент из заданных при вызове
        """
        #если нельзя использовать ничего, то ошибка
        if self.use_low == False and self.use_upper == False and self.use_simb == False and self.use_num == False:
            raise ValueError('не из чего генерировать')
        
        #если заданная длина пароля меньше длины пароля со всеми требуемыми символами
        elif self.length < (self.use_low + self.use_upper + self.use_simb + self.use_num):
            raise ValueError('заданная длина меньше требуемой')
        
        else: # все ок
            import random as rand
            
            el_for_template = self.template_el_gen() #получить один раз, не нужно вызывать каждый раз
            
            self.counter = 0 #счетчик на всякий случай
            
            #пока в шаблоне не будет представлен каждый требуемый элемент хотя бы один раз
            while set(el_for_template) != set(self.pass_template):
                self.pass_template = ''
                self.counter += 1
                for i in range(self.length):
                    #создаем шаблон пароля путем набора случайных элементов списка доступных элементов
                    self.pass_template += el_for_template[rand.randint(0,len(el_for_template)-1)]
                
                if self.counter > 100:
                    print('Более 100 итерация при генерации шаблона')
                    raise ValueError

            return self.pass_template
            
            
    def template_el_gen(self):
        """
        Генерация списка символов для последующей генерации пароля
        Список генерируется в соответствии с требованием наличия тех или иных видов элементов пароля
        """
        if self.use_low: self.el_for_template += 'l'
        if self.use_upper: self.el_for_template += 'u'
        if self.use_simb: self.el_for_template += 's'
        if self.use_num: self.el_for_template += 'n'
        return self.el_for_template
        
    
    def get_rand_low(self):
        """
        Возвращает случайный элемент строки из символов в нижнем регистра
        """
        import random as rand
        return self.usage_char[rand.randint(0,len(self.usage_char)-1)]
    
    
    def get_rand_upper(self):
        """
        Возвращает случайный элемент строки из символов в верхнем регистра
        """
        import random as rand
        return self.usage_char[rand.randint(0,len(self.usage_char)-1)].upper()
    
    
    def get_rand_simb(self):
        """
        Возвращает случайный элемент строки из специальных символов
        """
        import random as rand
        return self.usage_simb[rand.randint(0,len(self.usage_simb)-1)]
    
    
    def get_rand_num(self):
        """
        Возвращает случайный элемент строки из цифр
        """
        import random as rand
        return self.usage_num[rand.randint(0,len(self.usage_num)-1)]


a = MyPass(length = 4, use_num = True, use_low = True, use_upper = True, use_simb = True)
a.usage_simb += '-'
print('пароль:',a.pass_gen())
print('генерация с попытки:',a.counter)
```

```python
#Функция преобразования CamelCase в snake_case и kebab-case


def case_transform(s,out_type = 'snake'):
    """
    Принимает на вход строку в CamelCase и преобразует в snake или kebab
    Возвращает преобразованную строку или None, если не удалось это сделать
    """
    
    if ' ' in s or '_' in s or '-' in s: return #если есть пробел, тире или нижнее подчеркивание
    
    #преобразование в snake по-умолчанию
    new_s = ''
    for i, char in enumerate(s):
        if char == char.upper() and i != 0:
            new_s = new_s + '_' + char.lower()
        else:
            new_s += char.lower()
    
    if out_type == 'snake':
        return new_s
    elif out_type == 'kebab':
        return new_s.replace('_','-')

    
print(case_transform('outSplitTime','kebab'))
```
