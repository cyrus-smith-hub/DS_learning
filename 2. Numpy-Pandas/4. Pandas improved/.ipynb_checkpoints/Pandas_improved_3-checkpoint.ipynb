```python
import pandas as pd

log = pd.read_csv('visit_log.csv', sep=';')
```

```python
log.head()
```

```python
log.info()
```

```python
#Фильтрация с помощью loc
#Фильтр на строки : фильтр на столбцы
# : - выбор всех значений
log.loc[:,['visit_id', 'url']].head()
```

```python
#Фильтр на столбцы и строки

log.loc[log['region'] == 'Russia',['visit_id', 'url','region']].sample(10)
```

```python
#Фильтрация столбцов с помощью номера
log.iloc[:,[1,2]].head()
```

```python
#Выбор по строкам и столбцам
log.iloc[10:20,2:4].head()
```

```python
log.iloc[:,list(range(-1,-3,-1))].head()
```

```python
#Изменить порядок столбцов
log.iloc[:,list(range(-1,-len(log.keys()),-1))].head()
```

```python
#Фильтр с помощью loc

log.loc[log['region'] == 'Russia'].head()
```

```python
#Добавление cстолбца
log.loc[:,'new'] = 1
```

```python
log
```

```python
#Фильтр и действие с помощью loc
#Добавит столюец VAT и значения только если выполняется условие, иначе NaN

log.loc[log['region'] == 'Russia', 'VAT'] = 1.2
log.sample(10)
```

```python
#Lambda функции
#Здесь просто фильтр

print('before lambda')
print(log.count(),end='\n\n')

#попадет все, для чего лямбда = True
log.loc[lambda row: row['region'] == 'Russia'].info()
```

```python
#Вурнуть все через лямбда, просто пример
log.loc[lambda row: row.iloc[:,1] == row.iloc[:,1]]
```

```python
log.sample(5)
```

```python
%%time
#Еще фильтры, не через loc (медленно)
#Не забываем про axis = 1!!!

log['Russia_time_stamp_lambda'] = log.apply(lambda row: row['timestamp'] if 'Russia' in row['region'] else None, axis = 1)
log.sample(10)
```

```python
%%time
#Тоже самое но через loc

log['Russia_time_stamp_loc'] = log.loc[log['region'].str.contains('Russia'),'timestamp']
log.head()
```

```python
#isin
#только полное совпадение
#в ответе не будет яндекса

log.loc[log['traffic_source'].isin(['ya','google'])]
```

```python
#в то время как str.contains это проверка на содержание подстроки
#в ответе будет yandex
log.loc[log['traffic_source'].str.contains('ya|google')]
```

```python
#Пустые значения
import numpy as np

data = pd.DataFrame({'name':['Moscow', np.NAN, np.NAN, np.nan, 'others']})
data.head()
```

```python
#Фильтр на пустые

data.loc[pd.isnull(data['name'])]
```

```python
#наоборот
data.loc[~pd.isnull(data['name'])].head()
```
