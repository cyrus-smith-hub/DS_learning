```python
#Join

import pandas as pd

mov = pd.read_csv('ml-latest-small/movies.csv')
rat = pd.read_csv('ml-latest-small/ratings.csv')
```

```python
rat_pivot = pd.DataFrame(rat.pivot_table(index = 'movieId', columns = 'rating', 
                values = 'timestamp', aggfunc = 'count', fill_value = 0,margins = True)).reset_index()
```

```python
#если будет использован параметр on, нужно чтобы хотя бы в одной таблице это было индексом
#если без on, то нужно чтобы у обеих таблиц индекс был одинаковым столбцом, здесь - movieId

rat_pivot.set_index('movieId')

mov_rat = mov.join(rat_pivot.set_index('movieId'), how='outer').fillna(0)
mov_rat.sample(20)
```
