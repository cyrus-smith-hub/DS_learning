```python
#Удаление дубликатов

ratings = pd.read_csv('ratings_example.txt', sep = '\t')
ratings.head()
```

```python
movies = pd.read_csv('movies_example.txt', sep = '\t')
movies.head()
```

```python
#Merge
#в отличие от join параметр on указывает на столбцы в обоих датафреймах
merge_df = ratings.merge(movies, how = 'outer', on = 'movieId')
merge_df.head(10)
```

```python
#Удаление дубликатов
#можно добавить inplace, чтобы сразу удалить
merge_df.drop_duplicates(subset = 'movieId', keep = 'first')
```

```python
#Дубликаты по нескольким столбцам
#Удалят строки, у которых совпадают значения в указанных столбцах

merge_df.drop_duplicates(subset = ['movieId', 'timestamp'], keep = 'first')
```
