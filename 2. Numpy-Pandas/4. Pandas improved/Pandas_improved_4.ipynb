```python
#Pymystem

from pymystem3 import Mystem

m = Mystem()

a = 'рубли рублю рубить рубит'

m.lemmatize(a)
```

```python
import pandas as pd

data = pd.read_csv('keywords.csv')
```

```python
data.head()
```

```python
sample_data = pd.DataFrame(data.sample(50))
sample_data.head()
```

```python
def func(df):
    print(df['keyword'])
    return ''.join(m.lemmatize(df['keyword']))[:-1]


sample_data['lem'] = sample_data.apply(func,axis = 1)
```
