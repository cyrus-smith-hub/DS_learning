```python
#вероятность выпадения различных сумм при игре 2, 3 и 4 кубиками

import pandas as pd
import numpy as np
import seaborn as sns
from tqdm.notebook import tqdm #для прогресс бара
```

```python
cube = pd.Series([1,2,3,4,5,6])
count = 10**6

count_2_cube = {}
count_3_cube = {}
count_4_cube = {}
count_5_cube = {}

for _ in tqdm(range(count)):
    
    sum_2 = np.random.randint(0,7) + np.random.randint(0,7)
    sum_3 = sum_2 + np.random.randint(0,7)
    sum_4 = sum_3 + np.random.randint(0,7)
    sum_5 = sum_4 + np.random.randint(0,7)
    
    if sum_2 in count_2_cube:
        count_2_cube[sum_2] += 1
    else:
        count_2_cube.setdefault(sum_2,1)
        
    if sum_3 in count_3_cube:
        count_3_cube[sum_3] += 1
    else:
        count_3_cube.setdefault(sum_3,1)
        
    if sum_4 in count_4_cube:
        count_4_cube[sum_4] += 1
    else:
        count_4_cube.setdefault(sum_4,1)
        
    if sum_5 in count_5_cube:
        count_5_cube[sum_5] += 1
    else:
        count_5_cube.setdefault(sum_5,1)
        
```

```python
sns.relplot(data = pd.Series(count_2_cube).sort_index(), kind = 'line')
sns.relplot(data = pd.Series(count_3_cube).sort_index(), kind = 'line')
sns.relplot(data = pd.Series(count_4_cube).sort_index(), kind = 'line')
sns.relplot(data = pd.Series(count_5_cube).sort_index(), kind = 'line')
```
