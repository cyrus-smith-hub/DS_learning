```python
import numpy as np

l = ('10;'*10).split(';')
print(l)

a = np.array(range(20))
#print(a)

x = np.array([1,2,3,4,5,6])
y = np.array([2,3,4,5,6,7])

#Поэлементные операции
print(x+y)
print(x**y)
print(x/y, end='\n\n')

#arange
r = np.arange(10)
print(r)
r = np.arange(0,101,17)
print(r)
print(type(r))

#Генерация векторов и матриц
print('\n',np.full(10,2))
print(np.full(12, 'qq'))
#для матрицы 4x16
print(np.full([4,16],2))

#Конкатенациия. !данные в кортежах
a = np.full(10,1)
b = np.full(6,6)
c = np.concatenate((a,b))
print(c)

#Random и перемешивание
import random
a = np.arange(1,20)
random.shuffle(a)
print(a)

#Reshape
print('\nreshape')
a = np.arange(10)
a.shape
print(a.reshape(2,5))
print(a.reshape(5,2))

#Transpone
print('\ntransp')
a = np.arange(12)
a = a.reshape(2,6)
a = a.T
print(a)

#Ravel Разглаживание
print('\nravel')
a = np.arange(12)
a = a.reshape(2,2,3)
print(a)
a = a.ravel()
print(a)

#Reshape разглаживание
print('\nreshape разглаживание')
a = np.arange(12)
a = a.reshape(2,2,3)
print('вектор')
print(a.reshape(12))
print('матрица из 1 строки')
print(a.reshape(1,12))

#Матрицы
a = np.zeros(10)
print('\n', a)

a = np.eye(5)
print('\n',a)

a = np.diag([0,1,2,3])
print('\n',a)

a = np.diag(np.arange(0,10))
print('\n',a)

a = np.random.random(22)
print('\n',a)

#linspace
#делит диапазон на заданное количество отсчетов
#границы входят в итоговый диапазон
a = np.linspace(1,10,101)
print('\n',a)
print(len(a))
```
