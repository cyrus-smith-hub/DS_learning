```python
import pandas as pd
import numpy as np

data = pd.read_csv('data_samples/ml-25m/ratings.csv')
```

```python
data.head()
```

```python
#распределение оценок для пользователя 1
data[data['userId'] == 1]['rating'].value_counts()
```

```python
#Сводная таблица для всех

data.pivot_table(index = 'userId', columns = 'rating',values = 'timestamp', aggfunc = 'count',
                 fill_value = 0,margins = True).reset_index().tail()
```

```python
#целые оценки ставят чаще чем дробные
data['rating'].value_counts().reset_index().plot.scatter(x = 'index', y = 'rating')
```

```python
#Heatmap
#Проверка зависимости количества каждой оценки от параметра timestamp
#Разбивка времени на дипазоны, расчет относителного количества каждой оценки для каждого временного диапазона

sample_count = 50
range_step = (data['timestamp'].max() - data['timestamp'].min())/(sample_count)
min_range = data['timestamp'].min()

        
def return_hour(timestamp):
    return int((timestamp - min_range) // range_step)
    

data['timestamp_hour'] = data['timestamp'].apply(return_hour)

#data['timestamp_hour'].value_counts()
```

```python
data.head()
```

```python
data.groupby('timestamp_hour').agg({'rating':'count'})
```

```python
plot_data = data.pivot_table(index = 'timestamp_hour', columns = 'rating', 
                             values = 'timestamp', aggfunc = 'count', fill_value = 0).reset_index()
```

```python
plot_data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

plt.figure(figsize = (20,6))
sns.heatmap(plot_data, annot=False).invert_yaxis()
```
