```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('D:/DS_data_sets/vgsales.csv')
```

### Итоги
Два виде перменных: непрерывные и дискретные

Инструменты:
 - гистограмма - даже для непрерывных путем разбиения на диапазоны, для дискретных - value_counts
 - средняя с осторожностью
 - медиана
 - процентили - значение, меньше которого имеют "процентиль" часть выборки, например 20 процентиль равен 102 - значит значене меньшее 102 имеют 20 процентов выборки. Важно! проценили лучше проверять через определение реальной доли
 - bosplot - медиана, квартили, межквартильный размах
 - describe
 - отклонение - лучше обычное, а не стандартное (mad)
 - корреляция -- только для сравнения силы связи нескольких переменных, если связь между ними заранее известна

```python
#Barplot
df['Platform'].value_counts(normalize = True).plot(kind = 'bar')
```

```python
#SNS barplot
sns.histplot(df['Platform'].value_counts(normalize = True))
```

```python
#Hist
df['User_Count'].hist(bins = 12)
```

```python
#Деление на интервалы для гистограммы, для каждого значения определяется интерва, куда оно попадает
df['User_count_bin'] = pd.cut(df['User_Count'],12)
df['User_count_bin']
```

```python
#подсчет количества записей с каждым интервалом
df.groupby('User_count_bin')['User_Count'].count().plot(kind = 'bar', logy = True)
```

```python
#диперсия
((df['Year_of_Release'] - df['Year_of_Release'].mean())**2).sum()/(len(df['Year_of_Release'].dropna())-1)
```

```python
#СКО, получилось такое же как и в describe, так как применялось не просто len на все, а с учетом пропусков
#так и делает describe

(((df['Year_of_Release'] - df['Year_of_Release'].mean())**2).sum()/(len(df['Year_of_Release'].dropna())-1))**0.5
```

```python
#Обычное отклонение
(df['Year_of_Release'] - df['Year_of_Release'].mean()).abs().mean()
```
